#### 条形图bar

三类鸢尾花中的每一个特征的平均值绘制一个条形图，使用堆积条形图

```python
df[df['class']=='Iris-versicolor']
```

```python
fig, ax = plt.subplots(figsize=(6,6))
bar_width = .8
labels = [x for x in df.columns if 'length' in x or 'width' in x] #取四个需要的列名
ver_y = [df[df['class']=='Iris-versicolor'][x].mean() for x in labels]
vir_y = [df[df['class']=='Iris-virginica'][x].mean() for x in labels]
set_y = [df[df['class']=='Iris-setosa'][x].mean() for x in labels]

x = np.arange(len(labels)) #0，1，2，3
ax.bar(x, vir_y, bar_width, bottom=set_y, color='darkgrey') #bottom参数将该序列（vir_y）的y点最小值设置为其下面那个序列（set_y）的y点最大值，这样创建堆积条形图
ax.bar(x, set_y, bar_width, bottom=ver_y, color='white')
ax.bar(x, ver_y, bar_width, color='black')

ax.set_xticks(x + (bar_width/2)) #调整标签间隔
ax.set_xticklabels(labels, rotation=-70, fontsize=12); #顺时针旋转70度
ax.set_title('Mean Feature Measurement By Class', y=1.01)
ax.legend(['Virginica','Setosa','Versicolor']) #创建图例
```

# 4.建模和评估

## 4.1 Statsmodels

Statsmodels用于探索数据，估计模型，并运行统计检验的python包。在这里用它构建简单的线性回归模型，为setosa类中花萼长度和宽度之间的关系进行建模。

首先，通过散点图目测两者的关系。

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(df['sepal width'][:50], df['sepal length'][:50])
ax.set_ylabel('Sepal Length')
ax.set_xlabel('Sepal Width')
ax.set_title('Setosa Sepal Width vs. Sepal Length', fontsize=14, y=1.02)
```

似乎有正向的线性关系。使用statsmodels运行线性回归模型，预估关系的强度。

```python
import statsmodels.api as sm

y = df['sepal length'][:50]
x = df['sepal width'][:50]
X = sm.add_constant(x)

results = sm.OLS(y, X).fit()
print(results.summary())
```

是线性回归，sepal length = 2.6447 +  0.6909 x sepal width

R2=0.558 p值（Prob）是非常显著的。

绘制回归线：

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.plot(x, results.fittedvalues, label='regression line')
ax.scatter(x, y, label='data point', color='r')
ax.set_ylabel('Sepal Length')
ax.set_xlabel('Sepal Width')
ax.set_title('Setosa Sepal Width vs. Sepal Length', fontsize=14, y=1.02)
ax.legend(loc=2)
```

## 4.2 Scikit-learn

https://scikit-learn.org

建立在python科学栈的核心模块之上，也就是numpy,scipy,pandas和matplotlib。

覆盖的领域包括：分类、回归、聚类、降维、模型选择和预处理。

使用iris数据建立一个分类器，然后学习如何用scikit-learn的工具来评估模型。

在scikit-learn中打造机器学习模型，第一步是要理解数据应该如何构建：

- 数字型nxm维的矩阵X
- nx1维的因变量向量y，可以是数字（连续/离散），也可以是离散的字符串类型

然后将这些向量传递到指定的分类器的.fit()方法。

scikit-learn最大的好处，就是每个分类器都尽最大可能地使用同样地方法。

第一个例子：

```python
from sklearn.ensemble import RandomForestClassifier #随机森林分类器
from sklearn.model_selection import train_test_split #数据划分

clf = RandomForestClassifier(max_depth=5, n_estimators=10)

X = df.iloc[:,:4]
y = df.iloc[:,4]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

rf = pd.DataFrame(list(zip(y_pred, y_test)), columns=['predicted', 'actual'])
rf['correct'] = rf.apply(lambda r: 1 if r['predicted'] == r['actual'] else 0, axis=1)
    
rf
```

```python
rf['correct'].sum()/rf['correct'].count()
```

- 导入随机森林分类器
- 导入测试集划分
- 实例化分类器，选择了10个决策树的森林，每颗树允许五层的判定深度（避免过拟合）
- 创建X,y
- 测试集划分30%
- 训练模型.fit()
- 预测测试集结果

下面看看哪些特征提供了最佳的辨别力/预测能力

```python
f_importances = clf.feature_importances_
f_names = df.columns[:4]
f_std = np.std([tree.feature_importances_ for tree in clf.estimators_], axis=0)

zz = zip(f_importances, f_names, f_std)
zzs = sorted(zz, key=lambda x: x[0], reverse=True)

imps = [x[0] for x in zzs]
labels = [x[1] for x in zzs]
errs = [x[2] for x in zzs]

plt.bar(range(len(f_importances)), imps, color="r", yerr=errs, align="center")
plt.xticks(range(len(f_importances)), labels);
```

花瓣长宽度和长度具有较好的分辨力。这里包括了标准差，用于说明每个特征是多么的一致。生成方法：对于每个特征，获取10棵决策树的特征重要性，并计算标准差。

切换分类器，使用支持向量机（SVM）：

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

clf = OneVsRestClassifier(SVC(kernel='linear'))

X = df.iloc[:,:4]
y = np.array(df.iloc[:,4]).astype(str)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)

clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

rf = pd.DataFrame(list(zip(y_pred, y_test)), columns=['predicted', 'actual'])
rf['correct'] = rf.apply(lambda r: 1 if r['predicted'] == r['actual'] else 0, axis=1)
    
rf
```

```python
rf['correct'].sum()/rf['correct'].count()
```

在实例化分类器那里做了一个小小的格式的改变，因为SVM无法像随机森林那样，将这些标签解释为NumPy的字符串。

这些仅仅是scikit-learn能力的一小部分，但它应该可以说明这个伟大的工具对于机器学习应用而言强大的功能和力量。
